Estructuras de datos Python

Tuplas: Son inmutables (no se pueden modificar una vez asignados los valores)

Listas: Estructura de datos simples o compuestos, donde las casillas (posición) almacenan valores variables

```
#tupla
datos=(1,2,3)
print(datos)
#listas
estaturas=[1.56,2.01,1.75,1.5,1.79]
pesos=[60,89,78,54,81]

print(estaturas[0])
print(pesos[3])
```

```
for i in range(len(estaturas)):#repetir tantas veces como elementos haya en estaturas
    print(pesos[i]/estaturas[i]**2)
```

Las funciones en Python: se definen a partir de la palabra clave def  

Permiten crear bloques de código que se ejecutan cuando se llama la función. son la base de la modularidad de un programa en Python. Pueden ejecutarse sin pasar parámetros, o algunas requieren unos parámetros que se le pasan cunado se llama la función para poder ejecutar las operaciones con los valores que se le pasan a la función.

```
def imc(est,peso):
    return peso/est**2
```
